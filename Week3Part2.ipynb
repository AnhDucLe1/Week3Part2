```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import requests
```

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
r = requests.get(url, headers= 0)
tables = pd.read_html(r.text)
```

```python
df=pd.DataFrame(tables[0])
df.columns=['Postcode','Borough','Neighbourhood']

df.drop([0],axis=0,inplace=True)

df.reset_index()
```

```python
filtered_df = df[df.Borough != "Not assigned"]
filtered_df.shape
```

```python
filtered_df['New_Neighbourhood'] = np.where(filtered_df['Neighbourhood']=='Not Assigned', filtered_df['Borough'], filtered_df['Neighbourhood'])
filtered_df.head(10)
```

```python
postal_df = pd.DataFrame(filtered_df.groupby(['Postcode','Borough'])['New_Neighbourhood'].apply('.'.join).reset_index())
```

```python
postal_df = postal_df.rename(columns = {"Postcode": "PostalCode","New_Neighbourhood":"Neighbourhood"})
print(postal_df.head())
```

```python
postal_df.shape
```

```python
CanadaLatLon = pd.read_csv('https://cocl.us/Geospatial_data')
CanadaLatLon.head()
```

```python


CanadaLatLon.rename(columns ={'Postal Code':'PostalCode'}, inplace = True)
Canada_df = pd.merge(postal_df, CanadaLatLon, on='PostalCode')
Canada_df.head()

```

```python
Toronto_df = Canada_df[Canada_df['Borough'].str.contains('Toronto',regex = False)]
Toronto_df.head()
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
Toronto_df
```

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(Toronto_df['Latitude'],Toronto_df['Longitude'],Toronto_df['Borough'],Toronto_df['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

Using K = 6 for nearest 6 neighbour to cluster set 

```python
k = 6
Toronto_KmeanClust = Toronto_df.drop(['PostalCode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(Toronto_KmeanClust)
kmeans.labels_
Toronto_df.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
Toronto_df
```

```python
#Starting location
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, neighbourhood, cluster in zip(Toronto_df['Latitude'], Toronto_df['Longitude'], Toronto_df['Neighbourhood'], Toronto_df['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
